```python
mo_path = '../../MoModels/series_pump_test.mo'#'../../MoModels/secondary_pump_test.mo'#'../../MoModels/nb_hydr_static.mo'
fmu_name = mo_path.split(os.sep)[-1].split('.mo')[0]
res_dir = 'secondary_pump_test_res'#'./nb_hydr_static_res'
sudo_pwd = '63z64q92zza'
```

```python
mo_engine = Engine(mo_path, res_dir, sudo_pwd, 
							log_level = 'DEBUG')

sim_res_dir = mo_engine.simulate(start_time=0, final_time=3600,
					res_names = [
								 'pipe_6.m_flow', 'bypass_pipe.m_flow',
                                 'secdy_ret_pipe.m_flow'],
                    set_param_dict = {'chwp18_speed_const.k': 1, 
                                      'chwp1_speed_const.k': 1}
                                )
res_pd = pd.read_csv('{}{}{}_res.csv'.format(sim_res_dir, os.sep, fmu_name))
```

```python
res_pd
```

```python
first_m = []
second_m = []
for i in range(1, 11):
    secondary_speed = i/10
    sim_res_dir = mo_engine.simulate(start_time=0, final_time=3600,
					res_names = [
								 'pipe_6.m_flow', 'pipe_10.m_flow', 'bypass_pipe.m_flow',
                                 'secdy_ret_pipe.m_flow'],
                    set_param_dict = {'chwp18_speed_const.k': secondary_speed, 
                                      'chwp1_speed_const.k': 1}
                                )
    res_pd = pd.read_csv('{}{}{}_res.csv'.format(sim_res_dir, os.sep, fmu_name))
    first_m_i = res_pd['pipe_10.m_flow'].iloc[-1]
    second_m_i = res_pd['pipe_6.m_flow'].iloc[-1]
    first_m.append(first_m_i)
    second_m.append(second_m_i)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(first_m, color = 'k', label = 'first')
plt.plot(second_m, color = 'b', label = 'second')
plt.title('first fixed at 1, second speed changing')
plt.legend()
```
