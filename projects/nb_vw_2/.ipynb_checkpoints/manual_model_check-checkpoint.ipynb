```python
from HydraulicSimulator.HydraulicSimulator import Simulator
```

```python
mo_path = '../../../MoModels/nb_hydr_static_v4.mo'
fmu_path = './nb_hydr_static_sim_v4_res/compilation/nb_hydr_static_v4.fmu'
res_dir = './nb_hydr_static_sim_v4_res'
sudo_pwd = '63z64q92zza'
log_level = 'DEBUG'
```

```python
import pandas as pd
data_range_l = 100
data_range_h = 150
calibration_data = pd.read_csv('2021P2_calibration_data_30m.csv')[data_range_l:data_range_h]
calibration_data['p_diff'] = calibration_data['NB2_S_1_NYZ_sys_x_PcwOut_x']\
                            - calibration_data['NB2_S_1_NYZ_sys_x_PcwIn_x']
#calibration_data['NB2_S_1_NYZ_sys_x_PcwIn_x'] = 28000
sim_inputs = {'pump_speed_1.k': calibration_data['NB2_S_1_NYZ_cwp_9_HzSPR_x'].values/50,
			 'pump_speed_2.k': calibration_data['NB2_S_1_NYZ_cwp_10_HzSPR_x'].values/50,
			 'pump_speed_3.k': calibration_data['NB2_S_1_NYZ_cwp_11_HzSPR_x'].values/50,
			 'pump_speed_4.k': calibration_data['NB2_S_1_NYZ_cwp_12_HzSPR_x'].values/50,
             'val_pos_1.k': calibration_data['cwp_9_val_pos'],
             'val_pos_2.k': calibration_data['cwp_10_val_pos'],
             'val_pos_3.k': calibration_data['cwp_11_val_pos'],
             'val_pos_4.k': calibration_data['cwp_12_val_pos'],
             'ret_p.k': calibration_data['NB2_S_1_NYZ_sys_x_PcwIn_x'].values,
             'chw_sup_P.p': calibration_data['NB2_S_1_NYZ_sys_x_PcwOut_x'].values
			}
tgt_sup_pre = {
			 'tgt_sup_pre_pv_name': 'chw_sup_P.p',
			 'sup_pre_cv_name': 'terminal_resist_ratio.k',
			 'sup_pre_cv_range': [0.1, 1]}
outputs = ['chw_sup_m.m_flow', 'chw_sup_P.p', 'terminal_resist.y_actual', 'chw_term_P.p']
calibration_data
```

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(4,1,figsize=(15,10))
axs[0].plot(calibration_data['NB2_S_1_NYZ_cwp_9_HzSPR_x'])
axs[0].set_ylim([0, 50])
axs[1].plot(calibration_data['NB2_S_1_NYZ_cwp_10_HzSPR_x'])
axs[1].set_ylim([0, 50])
axs[2].plot(calibration_data['NB2_S_1_NYZ_cwp_11_HzSPR_x'])
axs[2].set_ylim([0, 50])
axs[3].plot(calibration_data['NB2_S_1_NYZ_cwp_12_HzSPR_x'])
axs[3].set_ylim([0, 50])


```

### Exp1

```python
mo_set_params = {}
simulator = Simulator(mo_path=mo_path, res_dir=res_dir, 
				sudo_pwd=sudo_pwd, log_level=log_level,
                      step_size = 1000,
                     fmu_path = None, mo_set_params = mo_set_params,
                      tgt_sup_pre = tgt_sup_pre, tgt_search_its = 20,
                      tgt_err_thres_perct = 1, search_fail_return = True
                     )
sim_res = simulator.simulate(sim_inputs, outputs, threads_n = 3)
```

```python
fig, axs = plt.subplots(2,1,figsize=(15,10))
axs[0].plot(calibration_data['NB2_S_1_NYZ_sys_x_PcwOut_x'].values, color = 'k', label = 'True pressure (Pa)')
axs[0].plot(sim_res[:, 1], color = 'g', label = 'Predicted pressure (Pa)')
axs[0].legend()
axs[1].plot(calibration_data['NB2_S_x_NYZ_x_x_Fcw_x'].values, color = 'k', label = 'True flow rate (kg/s)')
axs[1].plot(sim_res[:, 0], color = 'g', label = 'Predicted flow rate (kg/s)')

axs[1].legend()
```

```python
import numpy as np
np.savetxt('sim_res_2021_100-150.csv', sim_res)
```
